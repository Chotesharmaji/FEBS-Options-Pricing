```python
import pandas as pd
import numpy as np
import copy
```

```python
s = 100 # stock price
k = 80 # strike price
r = 0.07 # risk free rate
v = 0.10 # volatility
t = (105/365) # Time of maturity(year)
V = 0.01 # Square of Volatilty
ds = 10 # delta t
dt = (t/10)
N = 10
M = 20
S = 200 # MaxStock Price
```

```python
# Getting the constanyt values : Am, Bm, Cm
list_a = []
list_b = []
list_c = []
for m in range(0, 21):
    a = (V * m * m * dt - r * m * dt) / 2
    list_a.append(a)

    b = (1 - V * m * m * dt)
    list_b.append(b)

    c = (V * m * m * dt + r * m * dt) / 2
    list_c.append(c)



# For the initial list filled at time=T with Price=max(S-K,0)
list_s = []
for i in range (1,21):
  x = max((S - i*10 - k) , 0)
  list_s.append(x)
list_s.reverse()
z = (S-k)
list_s.append(z)

# Constant A
A= 1/1+(r*dt)


# For getting values for different periods
N=19
price = pd.DataFrame()
y=120.0
list_s_new = copy.copy(list_s)
for j in range(1,11):
  pr=[]
  pr.append(0)
  for i in range(1, N+1):
    option_price = A * ((float(list_a[i]) * list_s_new[i-1]) + (float(list_b[i]) * list_s_new[i]) + (float(list_c[i]) * list_s_new[i + 1]))
    pr.append(option_price)
  pr.append(z)
  list_s_new = pr
  price[f'Price_{j}']=pr

# reversing the columns and rows
price=price.iloc[::-1]
price[price.columns[::-1]]
```
